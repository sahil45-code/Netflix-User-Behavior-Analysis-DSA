```python
import pandas as pd

df = pd.read_csv('netflix_user_behavior.csv')  # name must match uploaded file
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by Genre and sum the watch hours
genre_watch = df.groupby('Genre')['Watch_Hours'].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=genre_watch.values, y=genre_watch.index, palette='coolwarm')
plt.title("Total Watch Hours per Genre")
plt.xlabel("Watch Hours")
plt.ylabel("Genre")
plt.show()
```

```python
user_watch = df.groupby('User_ID')['Watch_Hours'].sum().sort_values(ascending=False)

# Plot Top 5 Bingers
plt.figure(figsize=(8,4))
sns.barplot(x=user_watch.head(5).index.astype(str), y=user_watch.head(5).values, palette='magma')
plt.title("Top 5 Binge Watchers (Total Hours)")
plt.xlabel("User ID")
plt.ylabel("Total Watch Hours")
plt.show()
```

```python
# Define binge as session with more than 3 hours
df['Binge'] = df['Watch_Hours'] > 3

# Count binge sessions per user
binge_counts = df[df['Binge']].groupby('User_ID').size().sort_values(ascending=False)

# Plot top bingers
plt.figure(figsize=(8,4))
sns.barplot(x=binge_counts.head(5).index.astype(str), y=binge_counts.head(5).values, palette='plasma')
plt.title("Top 5 Users by Binge Sessions (>3 hrs)")
plt.xlabel("User ID")
plt.ylabel("Number of Binge Sessions")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Genre', y='Rating', data=df, palette='pastel')
plt.title("Ratings Distribution by Genre")
plt.xticks(rotation=45)
plt.ylabel("Rating")
plt.xlabel("Genre")
plt.show()
```
